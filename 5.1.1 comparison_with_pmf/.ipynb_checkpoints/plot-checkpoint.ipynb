```python
import numpy as np
import pandas as pd
# from matplotlib import pyplot as plt
import pylab as plt
from sklearn.cluster import KMeans
from scipy.stats import sem
from tqdm import tqdm
import os
from matplotlib import rc
rc('font', **{'family': 'serif', 'serif': ['Computer Modern'], 'weight': 'bold'})
rc('text', usetex=True)
```

```python
result_ed = pd.read_csv('collect_oboe_performance/results/regrets_oboe_ed_incremental.csv', index_col=0, header=0)
result_qr = pd.read_csv('collect_oboe_performance/results/regrets_oboe_qr_incremental.csv', index_col=0, header=0)
result_ed_with_mf = pd.read_csv('collect_oboe_performance/results/regrets_oboe_ed_incremental_with_mf.csv', index_col=0, header=0)
result_qr_with_mf = pd.read_csv('collect_oboe_performance/results/regrets_oboe_qr_incremental_with_mf.csv', index_col=0, header=0)
result_ed_no_time = pd.read_csv('collect_oboe_performance/results/regrets_oboe_ed_incremental_number_constrained.csv', index_col=0, header=0)
result_ed_no_time_with_mf = pd.read_csv('collect_oboe_performance/results/regrets_oboe_ed_incremental_with_mf_number_constrained.csv', index_col=0, header=0)

result_pmf = pd.read_csv('collect_pmf_performance/results/regrets_all_100_iterations.csv', index_col=0, header=0)
```

```python
errmtx_shape = pd.read_csv("../../oboe/automl/defaults/error_matrix.csv", index_col=0, header=0).shape
```

# plot regret curves

```python
result_ed_mean = []
result_ed_std = []
result_ed_ste = []

result_ed_no_time_mean = []
result_ed_no_time_std = []
result_ed_no_time_ste = []

result_qr_mean = []
result_qr_std = []
result_qr_ste = []

result_pmf_mean = []
result_pmf_std = []
result_pmf_ste = []

result_ed_with_mf_mean = []
result_ed_with_mf_std = []
result_ed_with_mf_ste = []

result_qr_with_mf_mean = []
result_qr_with_mf_std = []
result_qr_with_mf_ste = []

result_ed_no_time_with_mf_mean = []
result_ed_no_time_with_mf_std = []
result_ed_no_time_with_mf_ste = []

ranks_to_plot = []

initial_round = True

for item in result_ed.columns[:-1]:
    ed_single_rank = np.array(result_ed.loc[:, item])
    result_ed_mean.append(np.mean(ed_single_rank))
    result_ed_std.append(np.std(ed_single_rank))
    result_ed_ste.append(sem(ed_single_rank, axis=0, ddof=0))
    
    ed_no_time_single_rank = np.array(result_ed_no_time.loc[:, item])
    result_ed_no_time_mean.append(np.mean(ed_no_time_single_rank))
    result_ed_no_time_std.append(np.std(ed_no_time_single_rank))
    result_ed_no_time_ste.append(sem(ed_no_time_single_rank, axis=0, ddof=0))
    
    qr_single_rank = np.array(result_qr.loc[:, item])
    result_qr_mean.append(np.mean(qr_single_rank))
    result_qr_std.append(np.std(qr_single_rank))
    result_qr_ste.append(sem(qr_single_rank, axis=0, ddof=0))
    

    
    rank = [int(s) for s in item.split() if s.isdigit()][0]
    # The "+1" is due to the fact that we also include the entry that is predicted to be the best when computing regret
    pmf_single_rank = np.array(result_pmf.iloc[:, np.where([str(rank+1) in item for item in result_pmf.columns])[0][0]])
    result_pmf_mean.append(np.mean(pmf_single_rank))
    result_pmf_std.append(np.std(pmf_single_rank))
    result_pmf_ste.append(sem(pmf_single_rank, axis=0, ddof=0))
    if initial_round:
        result_ed_with_mf_mean.insert(0, np.mean(pmf_single_rank))
        result_ed_with_mf_std.insert(0, np.std(ed_single_rank))
        result_ed_with_mf_ste.insert(0, sem(ed_single_rank, axis=0, ddof=0))
        
        result_ed_no_time_with_mf_mean.insert(0, np.mean(pmf_single_rank))
        result_ed_no_time_with_mf_std.insert(0, np.std(ed_single_rank))
        result_ed_no_time_with_mf_ste.insert(0, sem(ed_single_rank, axis=0, ddof=0))
        
        result_qr_with_mf_mean.insert(0, np.mean(pmf_single_rank))
        result_qr_with_mf_std.insert(0, np.std(ed_single_rank))
        result_qr_with_mf_ste.insert(0, sem(ed_single_rank, axis=0, ddof=0))
    else:
        
        ed_with_mf_single_rank = np.array(result_ed_with_mf.loc[:, item])
        result_ed_with_mf_mean.append(np.mean(ed_with_mf_single_rank))
        result_ed_with_mf_std.append(np.std(ed_with_mf_single_rank))
        result_ed_with_mf_ste.append(sem(ed_with_mf_single_rank, axis=0, ddof=0))

        ed_no_time_with_mf_single_rank = np.array(result_ed_no_time_with_mf.loc[:, item])
        result_ed_no_time_with_mf_mean.append(np.mean(ed_no_time_with_mf_single_rank))
        result_ed_no_time_with_mf_std.append(np.std(ed_no_time_with_mf_single_rank))
        result_ed_no_time_with_mf_ste.append(sem(ed_no_time_with_mf_single_rank, axis=0, ddof=0))

        qr_with_mf_single_rank = np.array(result_qr_with_mf.loc[:, item])
        result_qr_with_mf_mean.append(np.mean(qr_with_mf_single_rank))
        result_qr_with_mf_std.append(np.std(qr_with_mf_single_rank))
        result_qr_with_mf_ste.append(sem(qr_with_mf_single_rank, axis=0, ddof=0))       
        
    initial_round = False        
    ranks_to_plot.append(rank)
    
ranks_to_plot = np.array(ranks_to_plot) + 1
```

plot Figure 5 (left), comparison among ED, QR and PMF:

```python
fontsize = 30
linewidth = 2
elinewidth = 1.5
with_error_bar = True
xtick_pos = np.linspace(5, 35, 4)
xtick_percentages = xtick_pos / errmtx_shape[1]
xticks_to_plot = ["{}({}\%)".format(int(xtick_pos[i]), int(xtick_percentages[i]*100)) for i in range(len(xtick_pos))]
colors = ['black', 'purple', 'orange', 'red', 'green', 'cyan', 'blue']

plt.figure(figsize=(8, 6))
plt.errorbar(ranks_to_plot, result_ed_no_time_mean, result_ed_no_time_ste, color=colors[0], label='ED (number)', elinewidth=elinewidth)
plt.errorbar(ranks_to_plot, result_ed_no_time_with_mf_mean, result_ed_no_time_with_mf_ste, color=colors[1], label='ED (number) with meta-features', elinewidth=elinewidth)
plt.errorbar(ranks_to_plot, result_pmf_mean, result_pmf_ste, color=colors[4], label='PMF', elinewidth=elinewidth)
plt.errorbar(ranks_to_plot, result_qr_mean, result_qr_ste, color=colors[2], label='QR', elinewidth=elinewidth)
plt.errorbar(ranks_to_plot, result_qr_with_mf_mean, result_qr_with_mf_ste, color=colors[3], label='QR with meta-features', elinewidth=elinewidth)
        
plt.xlabel('number (percentage) of observed entries', fontsize=fontsize)
plt.ylabel('regret (mean $\pm$ standard error)', fontsize=fontsize)
plt.legend(loc='upper center', bbox_to_anchor=(0.5, -0.2), fancybox=True, shadow=True, ncol=2, fontsize=fontsize-11)

plt.xticks(xtick_pos, xticks_to_plot, fontsize=fontsize)
plt.yticks(fontsize=fontsize)
plt.ylim(0, 0.035)
plt.locator_params(axis='y', nbins=8)
```

# regret ranking plot (Figure 5 right)

```python
num_observed_all = [int(s) for item in result_ed.columns[:-1] for s in item.split() if s.isdigit()]

ed_time_ranking = np.zeros((len(result_ed_no_time.index), len(num_observed_all)))
ed_time_with_mf_ranking = np.zeros((len(result_ed_no_time.index), len(num_observed_all)))
ed_number_ranking = np.zeros((len(result_ed_no_time.index), len(num_observed_all)))
ed_number_with_mf_ranking = np.zeros((len(result_ed_no_time.index), len(num_observed_all)))
qr_ranking = np.zeros((len(result_ed_no_time.index), len(num_observed_all)))
qr_with_mf_ranking = np.zeros((len(result_ed_no_time.index), len(num_observed_all)))
pmf_ranking = np.zeros((len(result_ed_no_time.index), len(num_observed_all)))

counter = 0

for idx, dataset_id in enumerate(result_ed_no_time.index):
    for idx_num_entries, num_observed in enumerate(num_observed_all):
        tmp1, tmp2, tmp3, tmp4, tmp5 = np.argsort(np.array([result_ed_no_time.iloc[idx, idx_num_entries], \
                                                            result_ed_no_time_with_mf.iloc[idx, idx_num_entries], result_qr.iloc[idx, idx_num_entries], result_qr_with_mf.iloc[idx, idx_num_entries], result_pmf.iloc[idx, idx_num_entries]]))
        if num_observed > 25:
            if tmp1 < tmp3:
                counter += 1
            else:
                counter += -1
        ed_number_ranking[idx, idx_num_entries] = tmp1
        ed_number_with_mf_ranking[idx, idx_num_entries] = tmp2
        qr_ranking[idx, idx_num_entries] = tmp3
        qr_with_mf_ranking[idx, idx_num_entries] = tmp4
        pmf_ranking[idx, idx_num_entries] = tmp5
        
        
ed_number_ranking_mean = np.average(ed_number_ranking, axis=0) + np.ones(len(num_observed_all))
ed_number_ranking_ste = sem(ed_number_ranking, axis=0, ddof=0)
ed_number_with_mf_ranking_mean = np.average(ed_number_with_mf_ranking, axis=0) + np.ones(len(num_observed_all))
ed_number_with_mf_ranking_ste = sem(ed_number_with_mf_ranking, axis=0, ddof=0)
pmf_ranking_mean = np.average(pmf_ranking, axis=0) + np.ones(len(num_observed_all))
pmf_ranking_ste = sem(pmf_ranking, axis=0, ddof=0)
qr_ranking_mean = np.average(qr_ranking, axis=0) + np.ones(len(num_observed_all))
qr_ranking_ste = sem(qr_ranking, axis=0, ddof=0)
qr_with_mf_ranking_mean = np.average(qr_with_mf_ranking, axis=0) + np.ones(len(num_observed_all))
qr_with_mf_ranking_ste = sem(qr_with_mf_ranking, axis=0, ddof=0)
```

```python
fontsize_axes = 30
linewidth = 2
elinewidth = 1.5


   
plt.figure(figsize=(8, 6))
#     plt.errorbar(num_observed_all, ed_number_ranking_mean, result_ed_ste, color=colors[5], label='ED (time)', elinewidth=elinewidth)
#     plt.errorbar(num_observed_all, result_ed_with_mf_mean, result_ed_with_mf_ste, color=colors[6], label='ED (time) with meta-features', elinewidth=elinewidth)
plt.errorbar(num_observed_all, ed_number_ranking_mean, ed_number_ranking_ste, color=colors[0], label='ED (number)', elinewidth=elinewidth)
plt.errorbar(num_observed_all, ed_number_with_mf_ranking_mean, ed_number_with_mf_ranking_ste, color=colors[1], label='ED (number) with meta-features', elinewidth=elinewidth)
plt.errorbar(num_observed_all, pmf_ranking_mean, pmf_ranking_ste, color=colors[4], label='PMF', elinewidth=elinewidth)
plt.errorbar(num_observed_all, qr_ranking_mean, qr_ranking_ste, color=colors[2], label='QR', elinewidth=elinewidth)
plt.errorbar(num_observed_all, qr_with_mf_ranking_mean, qr_with_mf_ranking_ste, color=colors[3], label='QR with meta-features', elinewidth=elinewidth)

plt.xlabel('number (percentage) of observed entries', fontsize=fontsize)
plt.ylabel('rank (mean $\pm$ standard error)', fontsize=fontsize)
plt.legend(loc='upper center', bbox_to_anchor=(0.5, -0.2), fancybox=True, shadow=True, ncol=2, fontsize=fontsize-11)

plt.xticks(xtick_pos, xticks_to_plot, fontsize=fontsize)
plt.yticks(fontsize=fontsize)
plt.ylim((1.5, 4))
plt.locator_params(axis='y', nbins=8)
```
